```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import pandas as pd
import random
from selenium.common.exceptions import StaleElementReferenceException
import time
import re
chrome_driver_path=Service("chromedriver.exe")
driver=webdriver.Chrome(service=chrome_driver_path)
URL="https://www.smartprix.com/mobiles/apple-brand"
driver.get(URL)
price=driver.find_elements(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[2]//span[@class="price"]')

pricesL=[]
load=driver.find_element(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[3]')

try:
    while load.is_displayed():
        load.click()
except StaleElementReferenceException:
    pass
name=driver.find_elements(By.XPATH,'//a[@class="name clamp-2"]/h2')
names=[i.text for i in name]
price=driver.find_elements(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[2]//span[@class="price"]')
prices=[p.text for p in price]
details=driver.find_elements(By.XPATH,'//*[@class="sm-feat specs"]')
```

```python
time.sleep(10)
for x in range(len(prices)):
    time.sleep(2)
    details[x].click()
    handles = driver.window_handles
    time.sleep(2)
    driver.switch_to.window(handles[1])
    time.sleep(2)
    if driver.find_elements(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[3]/ul[1]'):
        pricesL.append(driver.find_element(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[3]/ul[1]').text.replace('\n₹',':').replace('\n',','))
    else:
        pricesL.append("NOT")
    time.sleep(2)
    driver.close()
    driver.switch_to.window(handles[0]) 

len(name),len(prices),len(pricesL)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import pandas as pd
import random
from selenium.common.exceptions import StaleElementReferenceException
import time
import re
chrome_driver_path=Service("chromedriver.exe")
driver=webdriver.Chrome(service=chrome_driver_path)
driver.get('https://www.smartprix.com/mobiles/')
all_brands=driver.find_elements(By.XPATH, '//*[@class="sm-cats round scrl"]')
all_brands=[i.text for i in all_brands]
all_brands_name=all_brands[0]
all_brands_name=all_brands_name.split('\n')
brand=input(f"\n\n{all_brands_name}\n\nPlease enter brand name:___")
driver.find_element(By.XPATH,f'//*[@href="/mobiles/{brand}-brand"]').click()


pricesL=[]
load=driver.find_element(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[3]')
try:
    while load.is_displayed():
        load.click()
except StaleElementReferenceException:
    pass
name=driver.find_elements(By.XPATH,'//a[@class="name clamp-2"]/h2')
names=[i.text for i in name]
price=driver.find_elements(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[2]//span[@class="price"]')
prices=[p.text for p in price]
details=driver.find_elements(By.XPATH,'//*[@class="sm-feat specs"]')
time.sleep(4)
for x in range(len(prices)):
    time.sleep(3)
    details[x].click()
    handles = driver.window_handles
    time.sleep(2)
    driver.switch_to.window(handles[1])
    time.sleep(1)
    if driver.find_elements(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[3]/ul[1]'):
        pricesL.append(driver.find_element(By.XPATH,'//*[@id="app"]/main/div[1]/div[2]/div[3]/ul[1]').text.replace('\n₹',':').replace('\n',','))
    else:
        pricesL.append("NOT")
    time.sleep(1)
    driver.close()
    driver.switch_to.window(handles[0]) 

len(name),len(prices),len(pricesL)
```
